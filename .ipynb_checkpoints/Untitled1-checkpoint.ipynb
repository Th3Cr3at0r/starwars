```python
import requests
import json as js
```

```python
def getRec(json):
    if( json['next'] != None ):
        url = json['next']
        resp = requests.get(url)
        next=resp.json()
        json['results'].extend(next['results'])
        json['previous']=next['previous']
        json['next']=next['next']
        getRec(json)
    return json
```

```python
url = "https://swapi.dev/api/people/"
resp = requests.get(url)
resp
json=resp.json()
json=getRec(json)
json["results"]
#Creating json string
jsonString = js.dumps(json["results"], indent=4)
#print(jsonString)
print(type(jsonString))
```
