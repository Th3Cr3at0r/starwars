```python
import requests
import json as js
```

```python
def getRec(json):
    if( json['next'] != None ):
        url = json['next']
        resp = requests.get(url, verify = False)
        next=resp.json()
        json['results'].extend(next['results'])
        json['previous']=next['previous']
        json['next']=next['next']
        getRec(json)
    return json
```

```python
url = "https://swapi.dev/api/people/"

resp = requests.get(url, verify = False)
resp
peopleJson=resp.json()
peopleJson=getRec(peopleJson)
peopleJson["results"]

for id,result in enumerate(peopleJson["results"]):
    if( result["birth_year"] == "unknown" ):
        result["birth_year"] = "-1"
    else:
        byL = len(result["birth_year"])
        result["birth_year"] = ""+result["birth_year"][:byL-3]
    print("Year:"+result["birth_year"])

#Creating json string
jsonString = js.dumps(peopleJson["results"], indent=4)
print(jsonString)
#print(type(jsonString))
```

```python
def getAll(resource):
    url =""
    if(resource == "people"):
        url = "https://swapi.dev/api/people/"
    elif(resource == "films"):
        url = "https://swapi.dev/api/films"
    else:
        url =""
    if(url != ""):
        response = requests.get(url, verify = False)
        json=response.json()
        return getRec(json)
    return None
        
        
```

```python
print(getAll("films"))
```
